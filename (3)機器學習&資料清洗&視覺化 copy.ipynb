### 機器學習：多元線性迴歸分析
#### 目標預測值(y)：RE24(得分期望值/情境得分能力，Run Expectancy Matrix)
- 單⼀打席的RE24是這樣計算的：打席完成時球隊的RE - 打席開始前球隊的RE + 球隊得分。
- 透過選⼿的逐打席RE24，我們便能得知該選⼿在某打席為球隊貢獻了多少得分期望值。
- 累積逐打席的RE24，則意味著該選⼿（打者）整體⽽⾔約略為球隊創造過多少得分（或是投⼿省下多少失分）。

#### 自變數(x)：安打數、打點？


#### WPA(勝率增加值，Win Probability Added，WPA)

- 當一名選手的WPA很高，就代表他常常在關鍵時刻敲出能增加球隊勝率的一擊（或投手能在關鍵時刻解決打者）
- 簡單來說，就是心臟大顆

[參考說明](http://jackybaseball.blogspot.com/2020/03/win-expectancy-we.html)



## 一、簡單線性迴歸（y=w*X+b）

- Y：RE24
- X：安打

```python
# 先抓出安打與打數關係的散點圖

import matplotlib.pyplot as plt

x = new_df['RE24']
y = new_df['打點']

plt.scatter(x, y)
plt.xlabel('安打')
plt.ylabel('RE24')
plt.title('散點圖')
plt.show()
```

```python
# 劃出迴歸預測直線（同下一個跑偏重值）
# 盡量找出一條最接近散點圖的直線

w = 0
b = 0

def plot_pred(w,b):
    y_pred = w * x + b
    plt.plot(x , y_pred, label='預測線')
    plt.scatter(x, y, label='真實數據')
    plt.xlabel('安打')
    plt.ylabel('RE24')
    plt.title('散点图')
    plt.xlim(0,6)
    plt.ylim(0,6)
    plt.legend()
    plt.show()

plot_pred(0.5,-0.2) #可以自行帶入w,b的數值觀察預測線是否接近實際數據
```

```python
#模型訓練；算出權重值跟偏置值

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import pandas as pd

#設定x,y的資料
x = new_df['安打'].values.reshape(-1, 1) #資料進行 reshape，將其轉換成一列，因為 LinearRegression 模型期望 x 是二維數組。
y = new_df['WPA']

# train(訓練集)
# test(測試集)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2,random_state=3)


# 使用StandardScaler 來對資料進行標準化
# 將資料縮放成平均值為 0，標準差為 1 的分佈
# 確保模型在訓練的過程中能夠更容易收斂，並且避免某些特徵的值範圍過大對模型造成的不利影響

std_x = StandardScaler()
x_train = std_x.fit_transform(x_train)
x_test = std_x.transform(x_test)
std_y = StandardScaler()
y_train = std_y.fit_transform(y_train.to_numpy().reshape(-1, 1))
y_test = std_y.transform(y_test.to_numpy().reshape(-1, 1))
lr = LinearRegression()
lr.fit(x_train, y_train)
print('權重值：{}'.format(lr.coef_))   #算出每一個X值的權重，負值是負相關
print('偏置值：{}'.format(lr.intercept_)) #偏誤值B
```

```python
#進行資料測試
#查看預測值跟真實值的差別
#y_predict(預測值) / y_real(真實值)

y_predict = std_y.inverse_transform(lr.predict(x_test))
y_real = std_y.inverse_transform(y_test)
for i in range(10):
  print('預測值：{}，真實值：{}'.format(y_predict[i], y_real[i]))
```

```python
#平均方差越小越準確
#平均方差最小時的W與b值，就是最佳直線

from sklearn.metrics import mean_squared_error

merror = mean_squared_error(y_real, y_predict)
print('平均方差：{}'.format(merror))
```

```python
# 計算R-squared
from sklearn.metrics import r2_score

# 計算 R-squared
r2 = r2_score(y_real, y_predict)
print('R-squared：{}'.format(r2))
```

```python
# 繪製實際值跟預測值的圖觀察是否接近

import seaborn as sns
import matplotlib.pyplot as plt

# 假設 x 是實際值，y 是預測值

plt.scatter(x_test, y_test, color='black')
plt.plot(x_test, y_predict, color='red')  # 將實際值的線畫在圖上，方便比較
plt.xlabel('實際值')
plt.ylabel('預測值')
plt.title('實際值 vs. 預測值')
plt.show()
```

```python
# 繪製實際值跟預測值的圖觀察是否接近(換成點點而已)

import seaborn as sns
import matplotlib.pyplot as plt

# 假設 x 是實際值，y 是預測值

plt.scatter(x_test, y_test, color='black', label='實際值')
plt.scatter(x_test, y_predict, color='red', label='預測值')  # 將預測值以散點方式呈現
plt.xlabel('實際值')
plt.ylabel('預測值')
plt.title('實際值 vs. 預測值')
plt.legend()  # 顯示圖例，區分實際值和預測值的顏色
plt.show()
```

```python
# 設定x,y
# 分割測試集跟訓練集

from sklearn.model_selection import train_test_split
x = new_df[[ '打擊率', '打點']]
y = new_df["RE24"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()

# 查看資料筆數
x_len = len(x)
x_train_len = len(x_train)
x_test_len = len(x_test)

print(f'總資料：{x_len}'), print(f'訓練集：{x_train_len}'), print(f'測試集：{x_test_len}')
```

```python
# 算出權重值跟偏誤值

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import pandas as pd


std_x = StandardScaler()
x_train = std_x.fit_transform(x_train)
x_test = std_x.transform(x_test)
std_y = StandardScaler()
y_train = std_y.fit_transform(y_train.to_numpy().reshape(-1, 1))
y_test = std_y.transform(y_test.to_numpy().reshape(-1, 1))
lr = LinearRegression()
lr.fit(x_train, y_train)
print('權重值：{}'.format(lr.coef_))   #算出每一個X值的權重，負值是負相關
print('偏置值：{}'.format(lr.intercept_)) #偏誤值b
```

```python
#查看預測值跟真實值的差別
#這邊可以畫成線性圖比較兩條線的差別

y_predict = std_y.inverse_transform(lr.predict(x_test))
y_real = std_y.inverse_transform(y_test)
for i in range(10):
  print('預測值：{}，真實值：{}'.format(y_predict[i], y_real[i]))
```

```python
#平均方差越小越準確
#平均方差最小時的W與b值，就是最佳直線

merror = mean_squared_error(y_real, y_predict)
print('平均方差：{}'.format(merror))
```

```python
# 計算R-squared(有夠不准氣死)

from sklearn.metrics import r2_score

# 計算 R-squared
r2 = r2_score(y_real, y_predict)
print('R-squared：{}'.format(r2))
```
