### 線性迴歸分析

#### WPA(勝率增加值，Win Probability Added，WPA)

- 當一名選手的WPA很高，就代表他常常在關鍵時刻敲出能增加球隊勝率的一擊（或投手能在關鍵時刻解決打者）
- 簡單來說，就是心臟大顆

[參考說明](http://jackybaseball.blogspot.com/2020/03/win-expectancy-we.html)

### RE24（得分期望值/情境得分能力，Run Expectancy Matrix）
- 單⼀打席的RE24是這樣計算的：打席完成時球隊的RE - 打席開始前球隊的RE + 球隊得分。
- 透過選⼿的逐打席RE24，我們便能得知該選⼿在某打席為球隊貢獻了多少得分期望值。
- 累積逐打席的RE24，則意味著該選⼿（打者）整體⽽⾔約略為球隊創造過多少得分（或是投⼿省下多少失分）。

## 一、簡單線性迴歸（y=w*X+b）

- Y：預測安打數
- X：打數

```python
# 先抓出安打與打數關係的散點圖

import matplotlib.pyplot as plt

x = new_df['打數']
y = new_df['安打']

plt.scatter(x, y)
plt.xlabel('打數')
plt.ylabel('安打')
plt.title('散點圖')
plt.show()
```

```python
# 劃出迴歸預測直線（同下一個跑偏重值）
# 盡量找出一條最接近散點圖的直線

w = 0
b = 0

def plot_pred(w,b):
    y_pred = w * x + b
    plt.plot(x , y_pred, label='預測線')
    plt.scatter(x, y, label='真實數據')
    plt.xlabel('打數')
    plt.ylabel('安打')
    plt.title('散点图')
    plt.xlim(0,6)
    plt.ylim(0,6)
    plt.legend()
    plt.show()

plot_pred(0.5,-0.2)
```

```python
# 製作動態圖表，找出最適合的直線(目前沒用)

from ipywidgets import interact

interact(plot_pred, w=(-100,100,1), b=(-100,100,1))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import pandas as pd

#設定x,y的資料
x = new_df['打數'].values.reshape(-1, 1) #資料進行 reshape，將其轉換成一列，因為 LinearRegression 模型期望 x 是二維數組。
y = new_df['安打']

# train(訓練集)
# test(測試集)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2)


# 使用StandardScaler 來對資料進行標準化
# 將資料縮放成平均值為 0，標準差為 1 的分佈
# 確保模型在訓練的過程中能夠更容易收斂，並且避免某些特徵的值範圍過大對模型造成的不利影響

std_x = StandardScaler()
x_train = std_x.fit_transform(x_train)
x_test = std_x.transform(x_test)
std_y = StandardScaler()
y_train = std_y.fit_transform(y_train.to_numpy().reshape(-1, 1))
y_test = std_y.transform(y_test.to_numpy().reshape(-1, 1))
lr = LinearRegression()
lr.fit(x_train, y_train)
print('權重值：{}'.format(lr.coef_))   #算出每一個X值的權重，負值是負相關
print('偏置值：{}'.format(lr.intercept_)) #偏誤值B
```

```python
#進行資料測試
#查看預測值跟真實值的差別
#這邊可以畫成線性圖比較兩條線的差別
#y_predict(預測值)

y_predict = std_y.inverse_transform(lr.predict(x_test))
y_real = std_y.inverse_transform(y_test)
for i in range(10):
  print('預測值：{}，真實值：{}'.format(y_predict[i], y_real[i]))
```

```python
#平均方差越小越準確
#平均方差最小時的W與b值，就是最佳直線

from sklearn.metrics import mean_squared_error

merror = mean_squared_error(y_real, y_predict)
print('平均方差：{}'.format(merror))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 假設 x 是實際值，y 是預測值

plt.scatter(x_test, y_test, color='black')
plt.plot(x_test, y_predict, color='red')  # 將實際值的線畫在圖上，方便比較
plt.xlabel('實際值')
plt.ylabel('預測值')
plt.title('實際值 vs. 預測值')
plt.show()
```

```python
# 進行預測

# 假設有新的打數數據，存儲在 new_data 中
import numpy as np

new_data = np.array([[4]]) 
print('打數:',new_data)

# 將新數據進行標準化
new_data_standardized = std_x.transform(new_data)

# 使用模型進行預測
predicted_percentage = std_y.inverse_transform(lr.predict(new_data_standardized))

print('預測的安打數：', predicted_percentage[0])
```

```python
# 計算R-squared(有夠不准氣死)

from sklearn.metrics import r2_score

# 計算 R-squared
r2 = r2_score(y_real, y_predict)
print('R-squared：{}'.format(r2))
```

```python
# cost function 成本函數
# cost = (真實數據-預測值)**2

x = new_df['打數']
y = new_df['安打']


#先假設
w = 10
b = 0

y_pred = w * x + b
cost = (y - y_pred)**2 #計算出成本函數
cost.sum() / len(x)    #成本函數平均數
```

```python
#將上面的成本函數寫成def可以帶入不同的w,b值

def compute_cost(x, y, w, b):
    y_pred = w * x + b
    cost = (y - y_pred)**2 
    cost = cost.sum() / len(x)   
    return cost
```

```python
#嘗試帶入不同的w,b呼叫上面的def

compute_cost(x, y, 2, -3)
```

```python
# 在b=0的情況下，測試w=-100~100的成本函數

costs = []
for w in range(-100,101):
    cost = compute_cost(x, y, w, 0)
    costs.append(cost)

costs
```

```python
# 將costs畫成一張圖來觀察
import matplotlib.pyplot as plt

plt.plot(range(-100,101), costs)
plt.title('cost function b=0 w=-100~100')
plt.xlabel('w')
plt.ylabel('cost')
plt.show()
```
