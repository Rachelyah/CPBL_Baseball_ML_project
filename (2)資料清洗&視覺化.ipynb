```python
# 繪製長條圖，比較梁家榮選手跟各隊對戰的打擊率

import matplotlib
import matplotlib.pyplot as plt
#from matplotlib.font_manager import fontManager
#fontManager.addfont('/content/static/NotoSansTC-Regular.ttf')
#matplotlib.rc('font', family='Noto Sans TC')

listx = df['場次']
listy = df['打擊率']
plt.bar(listx, listy, width=0.5, color=['r','g','c','m','y'])
#plt.title("資訊程式課程選修人數")
#plt.xlabel("程式課程")
#plt.ylabel("選修人數")
plt.show()
```

```python
# 單獨抓出跟四隊對戰數據

#對戰富邦
fubon = df[df['對手'] == '悍']
fubon_svg_mean = fubon['打擊率'].mean()
fubon_svg_mean

#對戰兄弟
brothers = df[df['對手'] == '象']
brothers_svg_mean = brothers['打擊率'].mean()
brothers_svg_mean

#對戰龍
drogans = df[df['對手'] == '龍']
drogans_svg_mean = brothers['打擊率'].mean()
drogans_svg_mean

#對戰獅
lions = df[df['對手'] == '獅']
lions_svg_mean = brothers['打擊率'].mean()
lions_svg_mean

# 繪製長條圖，比較梁家榮選手跟各隊對戰的打擊率

import matplotlib.pyplot as plt

listx = ['fubon', 'brothers', 'drogans', 'lions']
listy = [fubon_svg_mean,brothers_svg_mean, drogans_svg_mean,  lions_svg_mean]
plt.title('by teams')
plt.bar(listx, listy, width=0.5, color=['r','g','c','m','y'])
plt.show()
```
